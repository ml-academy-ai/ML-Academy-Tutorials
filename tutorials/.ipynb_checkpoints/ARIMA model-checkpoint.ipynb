```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from typing import List, Tuple
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsforecast import StatsForecast
from statsforecast.models import ARIMA, AutoARIMA
import warnings
warnings.filterwarnings('ignore')
```

# Create a dataset

We create a dataset with a trend and partial autocorrelation to fit the ARIMA model

```python
# Set random seed
np.random.seed(42)

# AR(3) parameters
ar_params = [0.2, -0.01, 0.02]
n = 300
noise = np.random.normal(0, 0.05, size=n)

# Generate AR(3) base
y = [0, 0, 0]
for t in range(3, n):
    val = (
        ar_params[0] * y[t - 1] +
        ar_params[1] * y[t - 2] +
        ar_params[2] * y[t - 3] +
        noise[t]
    )
    y.append(val)

# Add linear trend
trend = 0.01 * np.arange(n)
y_trended = np.array(y) + trend

# Add irregular cycles
cycle = (
    0.1 * np.sin(0.1 * np.arange(n)) + 
    0.05 * np.sin(0.25 * np.arange(n) + np.random.uniform(0, 2 * np.pi)) +
    0.02 * np.sin(0.5 * np.arange(n) + np.random.uniform(0, 2 * np.pi))
)

# Add to the signal
y_final = y_trended + cycle

# Create DataFrame
df = pd.DataFrame({'value': y_final})
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['value'])
```

# Building ARIMA model

ARIMA model has 3 main parametes:
- d - the order of the differencing part
- p - the order of the autoregressive part
- q - the order of the moving average part
- ---

To find these parameters, we can use the following procedure:

**Step 1. Check for Stationarity to find d parameter**

**Why**: ARIMA requires a stationary time series (constant mean/variance over time).

**How:**

1. Plot the series: Look for trends or seasonality.

2. Use statistical tests:

ADF Test (Augmented Dickey-Fuller): p-value < 0.05  likely stationary.


**If not stationary:**
1. Difference the series (y_t - y_{t-1}), then re-test.
2. d = 1: First differencing is usually enough. Rarely, d = 2 is needed.

If stationary, proceed.

---
**Step 2. Determine p (auto-regressive component) using partial autocorrelation function.**

Plot the partical autocorrelation function and look for significant spikes at lag p, then sharp drop-off.

---
**Step 3. Determine q (the moving average component) using autocorrelation function**

Plot the autocorrelation function and look for significant spikes at lag q, then sharp drop-off.

### Step 1 - Differencing

```python
def difference(series: List[float], order: int = 1) -> List[float]:
    """
    Perform differencing on a time series to remove trend.

    Args:
        series: The original time series.
        order: The number of times to difference the series.

    Returns:
        A differenced version of the time series.
    """
    for _ in range(order):
        series = [series[i] - series[i - 1] for i in range(1, len(series))]
    return series
```

```python
# Step 1: Differencing
diffed = difference(df['value'], order=1)
```

```python
# Plot
plt.figure(figsize=(10, 4))
plt.plot(diffed)
plt.title('Differenced Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

The time series looks stationary, but we can also use Augmented Dickey-Fuller (ADF) Test to check it statistically. The Null hypothesis of this test is that the time series is non-stationary, so:
1. p > 0.05: Non-stationary
2. p < 0.05: Stationary

```python
result = adfuller(diffed)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

We see that the p-value is much lower than 0.05, so we reject the null hypothesis and say that the diffeerenced time series is stationary.

---
**Conclusion:** d=1

### Step 2 - Auto-regressive ARIMA parameter p

We will use the partial correlation plot to find the parameter p. You can read more about partial correlation here: https://online.stat.psu.edu/stat510/lesson/2/2.2.

**Note:** we apply the partial autocorrelation functionn on the stationary or differenced time series. Otherwise, the coefficients at the autoregressive compontents might capture the trend which we do NOT want to have.

```python
fig, ax = plt.subplots()
plot_pacf(diffed, lags=10, ax=ax)
plt.show()
```

From the plot we see that:
1. Lag=5 is statistically significant
2. Starting from lag 6 - we can cut

**Conclusion:** p=2

### Step 3 - Selecting parameter q - Moving average part

To select q, we use the autocorrelation plot

```python
fig, ax = plt.subplots()
plot_acf(diffed, lags=10, ax=ax)
plt.show()
```

We choose q = 1

# Using SARIMA to add seasonalities

Create a dataset that has seasonalities

```python
# Set random seed
np.random.seed(42)

# AR(3) parameters
ar_params = [0.2, -0.01, 0.02]
n = 300
noise = np.random.normal(0, 0.1, size=n)

# Generate AR(3) base
y = [0, 0, 0]
for t in range(3, n):
    val = (
        ar_params[0] * y[t - 1] +
        ar_params[1] * y[t - 2] +
        ar_params[2] * y[t - 3] +
        noise[t]
    )
    y.append(val)

# Add linear trend
trend = 0.01 * np.arange(n)
y_trended = np.array(y) + trend

# Add single seasonal component
seasonal_period = 7  # e.g. seasonality every 50 steps
seasonality = 0.3 * np.sin(2 * np.pi * np.arange(n) / seasonal_period)

# Add to signal
y_final = y_trended + seasonality

# Create DataFrame
df = pd.DataFrame({'value': y_final})
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['value'])
```

# Step 1 - Identify the seasonality lenght component m

```python
fig, ax = plt.subplots()
plot_acf(df['value'], lags=30, ax=ax)
plt.show()
```

Seems the tops of the cycles are at every 7-th lag, we can re-conform that by making differencing with this lag and plotting the autocorrelation again.

```python
series_df = df['value'].diff(7).dropna()
```

```python
fig, ax = plt.subplots()
plot_acf(series_df, lags=30, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots()
plot_pacf(series_df, lags=30, ax=ax)
plt.show()
```

Now, we can clearly see that out assumption is correct, so **m=7**

```python
m = 7
```

# Step 2 - Identify d-parameter

```python
series_df = df['value'].diff(1).dropna()
result = adfuller(series_df)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

Since p_value < 0.05, we can set d=1

# Step 3 - Identify D-parameter

```python
series_df = df['value'].diff(m).dropna()
result = adfuller(series_df)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

Since p_value < 0.05, we can set D=1

# Step 4 - Identify p, q, P and Q

```python
# Perform differencing and seasonal differencing
series_df = df['value'].diff(m).dropna()
```

```python
fig, ax = plt.subplots()
plot_pacf(series_df, lags=30, ax=ax)
plt.show()
```

From here, we can select p=0 and P=1

```python
fig, ax = plt.subplots()
plot_acf(series_df, lags=30, ax=ax)
plt.show()
```

From here, we can select q=0 and Q=1

### Fit the model and forecast

```python
# Define train/test split
n = 50
df_train, df_test = df[:-n], df[-n:]

# Write the parameters
m = 7
p = 0
d = 1
q = 0
P = 1
D = 1
Q = 1


# Fit SARIMA model
# Replace (p,d,q) and (P,D,Q,s) as needed  here is an example
sarima_model = ARIMA(
    order=(p, d, q),
    seasonal_order=(P, D, Q),
    season_length=m
)

results = sarima_model.fit(df_train['value'])

# Forecast
forecast = results.predict(h=n)
df_test['forecast_sarima'] = forecast['mean']

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df_train.index, df_train['value'], label='Train')
plt.plot(df_test.index, df_test['value'], label='Test')
plt.plot(df_test.index, df_test['forecast_sarima'], label='forecast_sarima')
plt.legend()
plt.title("SARIMA Forecast")
plt.show()
```
